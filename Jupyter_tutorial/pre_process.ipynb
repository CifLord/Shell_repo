```python
import lmdb
import pickle
def generate_lmdb(data, pathname: str):
    """
    atoms_list:: Can be either a list of atoms objects or list of Data objects
    """
    pathname = pathname + '.lmdb' if '.lmdb' not in pathname else pathname
    db = lmdb.open(
        pathname,
        map_size=1099511627 * 16,
        subdir=False,
        meminit=False,
        map_async=True,
    ) 
    txn = db.begin(write=True)        
    length=txn.stat()['entries']        
    txn.put(f"{length}".encode('ascii'), pickle.dumps(data, protocol=0))
    txn.commit()
    db.sync()
    db.close() 
```

```python

def lmdb_add_info(datas, pathname: str):
    """
    atoms_list:: Can be either a list of atoms objects or list of Data objects
    """
    pathname = pathname + '.lmdb' if '.lmdb' not in pathname else pathname
    db = lmdb.open(
        pathname,
        map_size=1099511627 * 16,
        subdir=False,
        meminit=False,
        map_async=True,
    ) 
    for data in datas:
        # del data[i,:].edge_index
        # del data[i,:].cell_offsets
        txn = db.begin(write=True)        
        length=txn.stat()['entries']        
        txn.put(f"{length}".encode('ascii'), pickle.dumps(data, protocol=0))
        txn.commit()
        db.sync()
    db.close() 
```

```python
def tail_lmdb(pathname: str):
    # Check if the '.lmdb' extension is already present in the pathname
    # If not, add it to the end
    pathname = pathname + '.lmdb' if '.lmdb' not in pathname else pathname

    # Open the LMDB database
    db = lmdb.open(
        pathname,
        map_size=1099511627 * 16,
        subdir=False,
        meminit=False,
        map_async=True,
    )

    # Begin a write transaction
    txn = db.begin(write=True)

    # Get the number of entries in the database and store it as a key-value pair
    length = txn.stat()['entries']
    print(length)
    txn.put("length".encode(), pickle.dumps(length, protocol=0))

    # Commit the transaction, sync the database to disk, and close it
    txn.commit()
    db.sync()
    db.close()
```

```python
from tqdm import tqdm
def out_fn(dataloader,model,pathname):

    model.eval()    
    with torch.no_grad():
           for i, batch in tqdm(
                           enumerate(dataloader),
                        total=len(dataloader)
            ):   
                batch=batch.to(DEVICE)
                model=model.to(DEVICE) 
                output=model(batch)
                output=output.cpu()
                num_atoms=[]
                for j in range(len(batch)):
                    num_atoms.append(batch[j].natoms.cpu()[0])
                
                split_indices = np.cumsum(num_atoms)[:-1]
                outputs=np.split(output,split_indices,axis=0)
                datas=[]  
                for j in range(len(batch)):
                    data=batch[j]
                    data.latent=outputs[j]
                    del data.edge_index
                    del data.cell_offsets
                    datas.append(data)
                if i >20:                     
                     pathname='Data/eoh_ff.lmdb'
                lmdb_add_info(datas,pathname)

                
                     
                

                
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * self.length + 2 * self.width

# Here we declare that the Square class inherits from the Rectangle class
class Square(Rectangle):
    def __init__(self, length,width,height):        
        super().__init__(length,width)
        self.height=height
    def vol(self,):
        return self.height*2+self.length
```
